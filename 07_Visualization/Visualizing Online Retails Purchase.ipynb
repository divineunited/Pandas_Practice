### Step 1. Import the necessary libraries

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import pandas as pd
import numpy as np
sns.set(color_codes=True) #overide maplot libs ugly colours.
mpl.rcParams['figure.figsize'] = [13, 8] #default figure size
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Online_Retail/Online_Retail.csv). 

### Step 3. Assign it to a variable called online_rt
Note: if you receive a utf-8 decode error, set `encoding = 'latin1'` in `pd.read_csv()`.

```python
online_rt = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Online_Retail/Online_Retail.csv', encoding = 'latin1')

online_rt.head(10)
```

### Step 4. Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK

```python
online_rt.groupby('Country').Quantity.sum().sort_values(ascending = False).iloc[1:11].plot('bar')

plt.title('Histogram of Quantity Ordered by Country')
plt.ylabel('Items Ordered')
plt.figure(figsize=(20,10))
```

### Step 5.  Exclude negative Quatity entries

```python
online_rt.shape
```

```python
online_rt = online_rt[online_rt['Quantity'] > 0]

online_rt.shape
```

### Step 6. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries

```python
# creating our data for our scatterplot
scatter_netherlands = online_rt[online_rt.Country == 'Netherlands'].groupby('CustomerID').sum()
scatter_eire = online_rt[online_rt.Country == 'EIRE'].groupby('CustomerID').sum()
scatter_germany = online_rt[online_rt.Country == 'Germany'].groupby('CustomerID').sum()

sns.scatterplot(x='Quantity', y='UnitPrice', data=scatter_netherlands)
plt.title('Netherlands')
```

```python
sns.scatterplot(x='Quantity', y='UnitPrice', data=scatter_eire)
plt.title('Ireland')
```

```python
sns.scatterplot(x='Quantity', y='UnitPrice', data=scatter_germany)
plt.title('Germany')
```
